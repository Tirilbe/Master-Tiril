# CPTAC PDAC Multi-Omics Harmonization

This notebook harmonizes the preprocessed transcriptome and proteome matrices to one shared gene set for downstream analyses.

## Inputs

- Transcriptome matrices: `rna_cancer_pre_harmonized.txt`, `rna_normal_pre_harmonized.txt`
- Proteome matrices: `proteome_cancer_pre_harmonized_with_header.txt`, `proteome_healthy_pre_harmonized_with_header.txt`

## Outputs

- Common gene lists: `multiomics_common_genes.txt`, `mRNA_common_genes.txt`
- Final RNA matrices: `rna_*_FINAL_FINAL*.txt`
- Final proteome matrices: `proteome_*_FINAL_FINAL*.txt`

## Pipeline

1. Load pre-harmonized transcriptome and proteome matrices.
2. Restrict all four matrices to one shared gene-symbol set.
3. Write the final common gene list.
4. Export FINAL_FINAL transcriptome and proteome matrices for downstream analyses.

```python
from pathlib import Path

from utils.multiomics_processing import (
    harmonize_multiomics_matrices,
    load_expression_matrix,
    save_final_multiomics_outputs,
)
```

## Setup




Notebook helper functions are imported from `Pipeline/1_Data_Preprocessing/utils/multiomics_processing.py`. Imports and configuration are kept together so the notebook can be rerun top-to-bottom with explicit paths.

```python
PROCESSED_DIR = Path("../../data/processed")
GENE_DIR = PROCESSED_DIR / "genelist"
TRANSCRIPTOME_DIR = PROCESSED_DIR / "transcriptome"
PROTEOME_DIR = PROCESSED_DIR / "proteome"

RNA_TUMOR_PRE = TRANSCRIPTOME_DIR / "rna_cancer_pre_harmonized.txt"
RNA_NORMAL_PRE = TRANSCRIPTOME_DIR / "rna_normal_pre_harmonized.txt"
PROTEOME_TUMOR_PRE = PROTEOME_DIR / "proteome_cancer_pre_harmonized_with_header.txt"
PROTEOME_NORMAL_PRE = PROTEOME_DIR / "proteome_healthy_pre_harmonized_with_header.txt"

OUT_GENE_LIST = GENE_DIR / "multiomics_common_genes.txt"
OUT_GENE_LIST_COMPAT = GENE_DIR / "mRNA_common_genes.txt"
OUT_RNA_TUMOR = TRANSCRIPTOME_DIR / "rna_cancer_FINAL_FINAL.txt"
OUT_RNA_NORMAL = TRANSCRIPTOME_DIR / "rna_normal_FINAL_FINAL.txt"
OUT_RNA_TUMOR_NO_HEADER = TRANSCRIPTOME_DIR / "rna_cancer_FINAL_FINAL_no_header.txt"
OUT_RNA_NORMAL_NO_HEADER = TRANSCRIPTOME_DIR / "rna_normal_FINAL_FINAL_no_header.txt"
OUT_PROTEOME_NORMAL_HEADER = PROTEOME_DIR / "proteome_healthy_FINAL_FINAL_with_header.txt"
OUT_PROTEOME_TUMOR_HEADER = PROTEOME_DIR / "proteome_cancer_FINAL_FINAL_with_header.txt"
OUT_PROTEOME_NORMAL_NO_HEADER = PROTEOME_DIR / "proteome_healthy_FINAL_FINAL.txt"
OUT_PROTEOME_TUMOR_NO_HEADER = PROTEOME_DIR / "proteome_cancer_FINAL_FINAL.txt"
```

## Step 1: Load Pre-Harmonized Matrices

```python
rna_tumor_df = load_expression_matrix(RNA_TUMOR_PRE)
rna_normal_df = load_expression_matrix(RNA_NORMAL_PRE)
proteome_tumor_df = load_expression_matrix(PROTEOME_TUMOR_PRE)
proteome_normal_df = load_expression_matrix(PROTEOME_NORMAL_PRE)
```

## Step 2: Harmonize Transcriptome And Proteome

```python
(
    rna_tumor_final,
    rna_normal_final,
    proteome_tumor_final,
    proteome_normal_final,
    final_genes,
) = harmonize_multiomics_matrices(
    rna_tumor_df,
    rna_normal_df,
    proteome_tumor_df,
    proteome_normal_df,
    out_gene_list=OUT_GENE_LIST,
    extra_gene_lists=[OUT_GENE_LIST_COMPAT],
)
```

## Step 3: Save Outputs

```python
save_final_multiomics_outputs(
    rna_tumor_final,
    rna_normal_final,
    proteome_tumor_final,
    proteome_normal_final,
    out_rna_tumor=OUT_RNA_TUMOR,
    out_rna_normal=OUT_RNA_NORMAL,
    out_rna_tumor_no_header=OUT_RNA_TUMOR_NO_HEADER,
    out_rna_normal_no_header=OUT_RNA_NORMAL_NO_HEADER,
    out_proteome_normal_header=OUT_PROTEOME_NORMAL_HEADER,
    out_proteome_tumor_header=OUT_PROTEOME_TUMOR_HEADER,
    out_proteome_normal_no_header=OUT_PROTEOME_NORMAL_NO_HEADER,
    out_proteome_tumor_no_header=OUT_PROTEOME_TUMOR_NO_HEADER,
)
```
